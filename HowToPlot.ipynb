# Color Scheme, Figure Style, etc. for SNAIL

In this notebook, you will find code on how to style your graphs including font, font size, color schemes, marker size, line width, etc.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Font and Font Size

```python
# sets the default font family for all text elements 
#in Matplotlib plots to a generic "sans-serif" type
plt.rcParams['font.family'] = 'sans-serif' 

#specifies arial as the preferred font for plots
plt.rcParams['font.sans-serif'] = ['Arial']

plt.rcParams['axes.titlesize'] = 14  # font size for plot titles
plt.rcParams['axes.labelsize'] = 12  # font size for x and y axis labels
plt.rcParams['xtick.labelsize'] = 10 # font size for x-axis tick labels
plt.rcParams['ytick.labelsize'] = 10 # font size for y-axis tick labels
plt.rcParams['legend.fontsize'] = 10 # font size for legend
```

## Examples:

```python
w = np.random.rand(5000) * 100
z = np.random.rand(5000) * 100
color_code_by = np.random.rand(5000) * 100

plt.figure()
plt.scatter(w, z, c=color_code_by, cmap='viridis', s=0.1, alpha=0.8, vmin=0,vmax=100)
# c tells you what data to color code by, cmap tells you the color 
# s tells you point size, alpha tells you transparency of the marker
# set approriate vmin & vmax values for what you are color coding by

# label your axes
plt.xlabel('W')
plt.ylabel('Z')

# include title if needed
plt.title('Color Scheme for Metallicity')

# show the color bar and label it
cbar = plt.colorbar()
cbar.set_label('Color Scale Value')

plt.show()
```

```python
plt.figure()
plt.scatter(w, z, c=color_code_by, cmap='inferno', s=0.1, alpha=0.8, vmin=0,vmax=14.5)

plt.xlabel('W')
plt.ylabel('Z')
plt.title('Color Scheme for Age')

cbar = plt.colorbar()
cbar.set_label('Color Scale Value')

plt.show()
```

```python
x1 = np.linspace(0, 10, 5)
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [5, 4, 3, 2, 6]
y4 = [1, 5, 2, 2, 6]
y5 = [3, 2, 4, 6, 9]

plt.plot(x1, y1, label='APOGEE', linestyle='-', color='k')
plt.plot(x1, y2, label='TESS', linestyle='-', color='r')
plt.plot(x1, y3, label='APOKASC2', linestyle='-', color='#FFE200')
plt.plot(x1, y4, label='APOK2', linestyle='-', color='c')
plt.plot(x1, y5, label='APOKASC3', linestyle='-', color='b')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plot Example')

# include a legend, you can use loc='' to specify where it is placed
plt.legend()

plt.show()
```

```python
data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30, density=True, color='grey') 
# 'bins' controls the number of bars (you may want to change this value), density=True normalizes the histogram such that the area under the histogram sums to 1

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Histogram Example')

plt.show()
```

```python
data1 = np.random.randn(1000)  
data2 = np.random.normal(2, 1.5, 800)

plt.hist(data1, bins=30, density=True, alpha=0.5, label='Dataset 1', color='r')
plt.hist(data2, bins=30, density=True,alpha=0.5, label='Dataset 2', color='#FFE200')
# when plotting multiple histograms, use an alpha value of about 0.5
# remember to use the assigned colors for the training sets / data


plt.title('Multi Histogram Example')
plt.xlabel('X')
plt.ylabel('Y')

# include a legend, doesn't have to be upper right
plt.legend(loc='upper right')

plt.show()
```
